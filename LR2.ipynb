# *linear regression to three variables (two independent variables and one dependent variable).*

With two independent variables, the linear regression model becomes:

```
 y = b0 + b1*x1 + b2*x2
```

Where:
*   y is the dependent variable.
*   x1 and x2 are the independent variables.
*   b0 is the intercept.
*   b1 and b2 are the coefficients for x1 and x2, respectively.




````python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
````

````python
data = {'x1': [1, 2, 3, 4, 5],
        'x2': [2, 3, 4, 5, 6],
        'y': [3, 5, 7, 9, 11]}
df = pd.DataFrame(data)
````

````python
X = df[['x1', 'x2']]
y = df['y']
````

````python
model = LinearRegression()
model.fit(X, y)
````

````python
b0 = model.intercept_
b1, b2 = model.coef_
````

````python
print(f"Intercept (b0): {b0}")
print(f"Coefficient for x1 (b1): {b1}")
print(f"Coefficient for x2 (b2): {b2}")
````

````python
new_data = {'x1': [6, 7], 'x2': [7, 8]}
new_df = pd.DataFrame(new_data)
predictions = model.predict(new_df)
````

````python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the original data
ax.scatter(df['x1'], df['x2'], df['y'], c='blue', marker='o', label='Original Data')

# Plot the regression plane
x1_range = np.linspace(df['x1'].min(), df['x1'].max(), 10)
x2_range = np.linspace(df['x2'].min(), df['x2'].max(), 10)
X1, X2 = np.meshgrid(x1_range, x2_range)
Y = b0 + b1 * X1 + b2 * X2
ax.plot_surface(X1, X2, Y, color='red', alpha=0.5)

# Set labels and title
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('3D Linear Regression')
ax.legend()
````
